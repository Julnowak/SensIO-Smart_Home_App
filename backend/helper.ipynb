```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings

df = pd.read_csv('anomaly_two.csv', parse_dates=['timestamp'])
```

```python
plt.plot(df['timestamp'], df['energy_consumption'], label='Energy Consumption')
```

```python
subset = df[df['energy_consumption'] < 1100][df['energy_consumption'] > 650]

plt.plot(subset['timestamp'], subset['energy_consumption'], color='blue', label='Data')
```

```python
df['anomaly'] = False
df.loc[(df['energy_consumption'] > 1100) | (df['energy_consumption'] < 650), 'anomaly'] = True
```

```python
df.to_csv('anomaly_two_marked.csv', index=False)
```
